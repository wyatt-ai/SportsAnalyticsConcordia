### Let's plot all Dimitrov's first serves where he won the point

```python
# filtering with query method
points = points_data.loc[(points_data["serverId"]=="ATPD875") & (points_data["scorerId"]=="ATPD875")]
```

```python
points
```

```python
serve_location = points.loc[(points["position"]=="bounce") & (points["order"]<=4)]
```

```python
def norm(val):
    if val <=0:
        val = val * -1
    return val
```

```python
deuce_serve_location = serve_location.loc[serve_location["court"]=="DeuceCourt"]

serves = deuce_serve_location.copy()

serves["x"] = deuce_serve_location["x"].apply(lambda x: norm(x))
serves["y"] = deuce_serve_location["y"].apply(lambda y: norm(y))
```

```python
serves.plot.scatter(x="x",y="y") # Needs to be put on a court
```
